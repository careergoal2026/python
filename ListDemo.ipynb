```python
#sum of array
def sumOfArray(N,input):
  sum=0
  for i in range(N):
    sum+=input[i]
  return sum
N=5
input=[1,2,3,4,5]
sumOfArray(N,input)
```

```python
def secondLargest(input_list):
  if len(input_list)<2:
    return -1
  else:
    non_dup= list(set(input_list)) # Removes duplicates
    non_dup.sort() # Sorts the list in-place
    return non_dup[-2]





input_list=[1]
secondLargest(input_list)
```

```python
#Q9. Shopping list items
def shop_list():

  shopping_list=[]
  while True:
    my_list=input().lower()
    if my_list=="end":
      break
    else:
      shopping_list.append(my_list)
  for items in shopping_list:
    print(items)
shop_list()
```

```python
#First Multiple
def firstMulti(ls,x):
  for i in ls:
    if i%x==0:
      return i
      break

  return -1

ls = [2, 1, 2, 5]
x  = 4
firstMulti(ls,x)
```

```python
#Consecutive Duplicate
def cons_dup(ls):
  i=1
  for i in range(len(ls)):
    if ls[i]==ls[i-1]:
      return True

  return False

ls=[1,2,3,3]
cons_dup(ls)
```

```python
# Difference of even and odd
def diff(ls):
  sum_even=0
  sum_odd=0
  for no in ls:
    if no%2==0:
      sum_even+=no
    else:
      sum_odd+=no
  return sum_even-sum_odd
ls=[2 ,8 ,4 ,6]
diff(ls)
```

```python
#Divisible by 5 and 7
def div(ls):
  new_list=[]
  for i in range(len(ls)):
    if ls[i]%5==0 and ls[i]%7==0:
      new_list.append(ls[i])
  return new_list
ls=[5, 7, 70, 50, 35]
div(ls)
```

```python
#Product of elements
def product(ls):
  mul=1
  for i in range(len(ls)):
    mul*=ls[i]
  return mul

ls=[7, 9, 2, 51]
product(ls)
```
